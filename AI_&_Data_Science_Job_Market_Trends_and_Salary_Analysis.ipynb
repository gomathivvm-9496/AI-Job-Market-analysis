## Stage 2 – Data Cleaning and Pre-processing



```python
#Missing Values Check
df.isnull().sum()
```

```python
#Duplicate Check
df.duplicated().sum()
```

```python
#Rename column
df.rename(columns={
    "skills_python":"python_required",
    "skills_sql":"sql_required",
    "skills_ml":"ml_required",
    "skills_deep_learning":"deep_learning_required",
    "skills_cloud":"cloud_required",
    "job_posting_month":"posting_month",
    "job_posting_year":"posting_year"},inplace=True)
df.head()
```

```python
df.columns
```

```python
#outliers
num_cols=df.select_dtypes(include=["int64","float64"]).columns
num_cols
```

```python
import matplotlib.pyplot as plt

for col in num_cols:
  plt.figure()
  plt.boxplot(df[col])
  plt.title(col)
  plt.show()
```

# **Insights**
* Salary outliers were present in the dataset.
* Since these represent real high-paying jobs, I decided to retain them instead of removing them, as they provide important insights into salary distribution.

```python
#skew
df[["salary","years_experience","job_openings"]].skew()
```

# **Insights**
* Skewness was checked for numerical columns such as salary, years of experience, and job openings.
* The salary column was found to be right-skewed due to the presence of high salary values.
* Salary is positively skewed due to high-paying outliers.
* Therefore, no transformation was required

```python
df.dtypes
```

```python
#convert data types
df[["python_required","sql_required","ml_required","deep_learning_required","cloud_required"]]=\
df[["python_required","sql_required","ml_required","deep_learning_required","cloud_required"]].astype(bool)
```

```python
df.dtypes
```

```python
#change data type
df["posting_date"]=pd.to_datetime(df["posting_year"].astype(str)+"-"+df["posting_month"].astype(str),format="%Y-%m")
df.head()
```

```python
#drop column
df.drop(["posting_month","posting_year"],axis=1,inplace=True)
df.head()
```

# **Feature column**

```python
df["exp_category"]=pd.cut(df["years_experience"],bins=[0,2,5,10,20],labels=["Entry","Mid","Senior","Expert"],include_lowest=True)
df.head()
```

```python
df[["experience_level","exp_category"]].value_counts()
```

```python
df["salary_category"]=pd.cut(df["salary"],bins=[0,50000,100000,200000,500000],labels=["Low","Medium","High","very High"])
df.head()
```

```python
df["total_skills"]=(df["python_required"].astype(int)+
                    df["sql_required"].astype(int)+
                    df["ml_required"].astype(int)+
                    df["deep_learning_required"].astype(int)+
                    df["cloud_required"].astype(int))
df.head()
```

```python
df["is_remote"]=df["remote_type"].apply(lambda x:1 if x=="Remote" else 0)
df.head()
```

## Stage 3 – EDA and Visualizations



# **UNIVARIATE ANALYSIS (Single Variable)**

# **Job Role Distribution**

```python
import matplotlib.pyplot as plt
import seaborn as sns
df["job_title"].value_counts().head(10).plot(kind="bar")
plt.title("Top job Roles")
plt.xticks(rotation=45)
plt.show()
```

This chart is a bar plot that shows the count (frequency) of different job roles in the dataset.
Each bar represents a job title, and its height indicates how many times that role appears
# features:
* Column used: job_title
* Function used: value_counts()
* Chart type: Bar chart
# Interpretation
* The chart shows that all job roles have almost similar counts.
* Roles like Business Analyst, AI Engineer, and Machine Learning Engineer have slightly higher counts.
* There is no huge difference between roles, meaning demand is fairly balanced across job types.



# **Salary Distribution**

```python
plt.hist(df["salary"],bins=20)
plt.title("salary distribution")
plt.show()
```

* This chart is a histogram that shows the distribution of salaries in the dataset.
* It groups salary values into ranges (bins) and displays how many jobs fall into each range.
## Interpretation
* Most salaries are concentrated between 80,000 and 140,000
* The distribution looks approximately normal (bell-shaped)
* Very high salaries (above 180,000) are less frequent
* Very low salaries are also fewer

# features:
* Column used: salary
* Chart type: Histogram
* Function used: plt.hist()



# **Salary Boxplot (Outliers)**

This boxplot shows the distribution of salary values

# **BIVARIATE ANALYSIS (Two Variables)**

# **Experience vs Salary**

```python
df[["years_experience","salary"]].corr
sns.scatterplot(df,x="years_experience",y="salary")
plt.title("Experience vs salary")
plt.show()
```

This chart shows the relationship between years of experience and salary.
# Features:
* Columns used:
   * years_experience
   * salary
* Chart type: Scatter plot
* Library: Seaborn
# Interpretation:
* The scatter plot indicates a positive relationship between years of experience and salary. As experience increases, salary generally tends to increase.

* However, the relationship is not perfectly linear, as there is noticeable variation in salary at each experience level. This suggests that factors such as skills, job role, company, and location also play a significant role in determining salary.

* Additionally, some jobs with lower experience levels offer relatively high salaries, indicating the presence of high-demand roles or specialized skill requirements.
